```python
import pandas as pd
import numpy as np
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
```

```python
import seaborn as sns; sns.set(color_codes=True)
iris = sns.load_dataset("iris")
species = iris.pop("species")   # Remove the species column
```

```python
species.nunique()
```

```python
iris.shape
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
from sklearn.cluster import AgglomerativeClustering
```

Single Link Clustering

```python
# Apply Agglomerative Clustering
# Let's assume we want to create 3 clusters (as we know Iris has 3 species)
agg_clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
clusters = agg_clustering.fit_predict(iris)

# Plotting
plt.figure(figsize=(10, 7))
plt.scatter(iris.iloc[:, 0], iris.iloc[:, 1], c=clusters, cmap='rainbow')
plt.title('Agglomerative Clustering of Iris Dataset (Single Link)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Hierarchical Clustering
linked = linkage(iris, method='single')

# Plot the Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Iris dataset)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

```python
# Silhouette Score
silhouette_avg = silhouette_score(iris, clusters)
print("The average silhouette_score is :", silhouette_avg)

# Calinski-Harabasz Index
calinski_harabasz = calinski_harabasz_score(iris, clusters)
print("The Calinski-Harabasz Index is :", calinski_harabasz)

# Davies-Bouldin Index
davies_bouldin = davies_bouldin_score(iris, clusters)
print("The Davies-Bouldin Index is :", davies_bouldin)
```
